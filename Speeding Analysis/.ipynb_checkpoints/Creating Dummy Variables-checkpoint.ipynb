### 'ALCOHOL' is frequently mis-entered in the dataset

There are ~2,500 entries where df['Alcohol']==1. \
There are ~60,000 entries where "alcohol" is mentioned in df['Description']

```python
counter = 0
for i in range(len(df)):
    if df['Alcohol'].iloc[i] == 0:
        if type(df['Description'].iloc[i]) == str:
            if 'ALCOHOL' in df['Description'].iloc[i]:
                counter+=1
print(counter)
```

#### Fixing The 'ALCOHOL' Problem

Note: need to make all citations of a stop having Alcohol==1 if any Alcohol==1

```python
for i in range(len(df)):
    if df['Alcohol'].iloc[i] == 0:
        if type(df['Description'].iloc[i]) == str:
            if 'ALCOHOL' in df['Description'].iloc[i]:
                df['Alcohol'].iloc[i] = 1
```

```python
def homogenize_entries(dataframe, var: str, default, new) -> list:
    """
    var == column in dataframe
    default == default entry (usually "no")
    new == atypical entry (usually "yes")
    
    homogenize_entries('Alcohol', 0, 1) will make all citations of a stop have dataframe['Alcohol']==1 
    if ANY of them already have that.
    
    
    """
    
    
    index_dict = {ind:0 for ind in dataframe.index}
    for x in list(zip(dataframe.index,dataframe[var])):
        if x[1]==new:
            index_dict[x[0]] = 1
    
    new_col = list(dataframe[var])
    
    for i, x in enumerate(list(zip(dataframe.index,new_col))):
        if index_dict[x[0]]==1:
            new_col[i] = new
        
    return new_col        
```

```python
df.insert(0, 'Test Alcohol', homogenize_entries(df, 'Alcohol', 0, 1))
```
