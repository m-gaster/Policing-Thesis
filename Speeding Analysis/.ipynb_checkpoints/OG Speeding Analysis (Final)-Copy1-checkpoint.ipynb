```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
from datetime import time
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('tableau-colorblind10')

pd.set_option("display.max_rows", 50)
pd.options.display.width=None

df = pd.read_csv("Stops for Speeding Analysis.csv", dtype={'latitude': float, 'Longitude':float})

race_set = set(x for x in df["Race"])
```

```python
def speed_parser(string):
    if 'mph' in string or 'MPH' in string:
        temp_holder = []
        for word in string.split():
            if word.isdigit():
                temp_holder.append(int(word))
        if len(temp_holder) == 2:
            return sorted(temp_holder)
        elif len(temp_holder) == 1:
            return [temp_holder[0], np.nan]
        else:
            return [np.nan, np.nan]
    else:
        return [np.nan, np.nan]


Speed_Limit = [np.nan] * len(df)
Driving_Speed = [np.nan] * len(df)
Speed_over_posted_limit = [np.nan] * len(df)

for index, x in enumerate(df['Description']):
    if type(x) == str:
        temp = speed_parser(x)
        Speed_Limit[index] = temp[0]
        Driving_Speed[index] = temp[1]
        if Driving_Speed[index] != np.nan:
            Speed_over_posted_limit[index] = Driving_Speed[index]-Speed_Limit[index]
        else:
            Speed_over_posted_limit[index] = np.nan
    else:
        Speed_over_posted_limit[index] = np.nan
        
for i,x in enumerate(Speed_over_posted_limit):
    if x>100:
        Speed_over_posted_limit[i] = np.nan
```

```python
df.insert(5, 'Speed Limit', Speed_Limit)

df.insert(6, 'Driving Speed', Driving_Speed)

df.insert(7, 'Speed Over Posted Limit', Speed_over_posted_limit)
```

```python
white_speeding_dict = {}
non_white_speeding_dict = {}
black_speeding_dict = {}
hispanic_speeding_dict = {}
asian_speeding_dict = {}
native_american_speeding_dict = {}
other_speeding_dict = {}

for x in range(0,100):
    white_speeding_dict[x] = 0
    non_white_speeding_dict[x] = 0
    black_speeding_dict[x] = 0
    hispanic_speeding_dict[x] = 0
    asian_speeding_dict[x] = 0
    native_american_speeding_dict[x] = 0
    other_speeding_dict[x] = 0
    
for i in range(len(df)):
    if not np.isnan(df['Speed Over Posted Limit'].iloc[i]):
        if df['Race'].iloc[i] == 'WHITE':
            white_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
        else:
            non_white_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
            if df['Race'].iloc[i] == 'BLACK':
                black_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
            elif df['Race'].iloc[i] == 'HISPANIC':
                hispanic_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
            elif df['Race'].iloc[i] == 'ASIAN':
                asian_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
            elif df['Race'].iloc[i] == 'NATIVE AMERICAN':
                native_american_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
            elif df['Race'].iloc[i] == 'OTHER':
                other_speeding_dict[df['Speed Over Posted Limit'].iloc[i]] += 1
```

```python
"""
Setting up histograms for speeding among races
"""


n_groups = len(black_speeding_dict)
sum_of_white_speeding = sum(white_speeding_dict[x] for x in white_speeding_dict)
sum_of_black_speeding = sum(black_speeding_dict[x] for x in black_speeding_dict)
white_speeding_as_percent = [white_speeding_dict[x]/sum_of_white_speeding for x in white_speeding_dict]
black_speeding_as_percent = [black_speeding_dict[x]/sum_of_black_speeding for x in black_speeding_dict]

n_groups = len(hispanic_speeding_dict)
sum_of_white_speeding = sum(white_speeding_dict[x] for x in white_speeding_dict)
sum_of_hispanic_speeding = sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict)
white_speeding_as_percent = [white_speeding_dict[x]/sum_of_white_speeding for x in white_speeding_dict]
hispanic_speeding_as_percent = [hispanic_speeding_dict[x]/sum_of_hispanic_speeding for x in hispanic_speeding_dict]

n_groups = len(asian_speeding_dict)
sum_of_white_speeding = sum(white_speeding_dict[x] for x in white_speeding_dict)
sum_of_asian_speeding = sum(asian_speeding_dict[x] for x in asian_speeding_dict)
white_speeding_as_percent = [white_speeding_dict[x]/sum_of_white_speeding for x in white_speeding_dict]
asian_speeding_as_percent = [asian_speeding_dict[x]/sum_of_asian_speeding for x in asian_speeding_dict]

n_groups = len(native_american_speeding_dict)
sum_of_white_speeding = sum(white_speeding_dict[x] for x in white_speeding_dict)
sum_of_native_american_speeding = sum(native_american_speeding_dict[x] for x in native_american_speeding_dict)
white_speeding_as_percent = [white_speeding_dict[x]/sum_of_white_speeding for x in white_speeding_dict]
native_american_speeding_as_percent = [native_american_speeding_dict[x]/sum_of_native_american_speeding for x in native_american_speeding_dict]

n_groups = len(other_speeding_dict)
sum_of_white_speeding = sum(white_speeding_dict[x] for x in white_speeding_dict)
sum_of_other_speeding = sum(other_speeding_dict[x] for x in other_speeding_dict)
white_speeding_as_percent = [white_speeding_dict[x]/sum_of_white_speeding for x in white_speeding_dict]
other_speeding_as_percent = [other_speeding_dict[x]/sum_of_other_speeding for x in other_speeding_dict]
```

```python
"""
Plotting all races' clustering
"""

n_groups = len(white_speeding_as_percent)
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 1/len(race_set)
opacity = 1

rects1 = plt.bar(index, white_speeding_as_percent, bar_width,
alpha=opacity,
label='White')

rects6 = plt.bar(index + bar_width, asian_speeding_as_percent, bar_width,
alpha=opacity,
label='Asian')

rects2 = plt.bar(index + bar_width*2, other_speeding_as_percent, bar_width,
alpha=opacity,
label='Other')

rects3 = plt.bar(index + bar_width*3, hispanic_speeding_as_percent, bar_width,
alpha=opacity,
label='Hispanic')

rects4 = plt.bar(index + bar_width*4, black_speeding_as_percent, bar_width,
alpha=opacity,
label='Black')

rects5 = plt.bar(index + bar_width*5, native_american_speeding_as_percent, bar_width,
alpha=opacity,
label='Native American')





plt.xlabel('MPH Over Posted Limit')
plt.ylabel("% of own Racial Group's Stops")
plt.xlim(7,25)
plt.title('Distribution of Reported Speeding Infractions')
plt.legend(loc=1, prop={'size': 8})
# plt.grid(True)
plt.grid(linestyle='dotted')
style.use('tableau-colorblind10')
plt.savefig('Speeding Analysis Figure.png',dpi=1200)
```

```python
"""
Hypothesis testing for equal means
"""

standard_errors_dict = {}

for race in race_set:
    standard_errors_dict[race] = 0
    
standard_errors_dict['ASIAN'] = np.sqrt(((asian_speeding_dict[9]/sum(asian_speeding_dict[x] for x in asian_speeding_dict))*
                (sum(asian_speeding_dict[x] for x in asian_speeding_dict if x!=9)/
                sum(asian_speeding_dict[x] for x in asian_speeding_dict)))/
                sum(asian_speeding_dict[x] for x in asian_speeding_dict))

standard_errors_dict['BLACK'] = np.sqrt(((black_speeding_dict[9]/sum(black_speeding_dict[x] for x in black_speeding_dict))*
                (sum(black_speeding_dict[x] for x in black_speeding_dict if x!=9)/
                sum(black_speeding_dict[x] for x in black_speeding_dict)))/
                sum(black_speeding_dict[x] for x in black_speeding_dict))

standard_errors_dict['HISPANIC'] = np.sqrt(((hispanic_speeding_dict[9]/sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict))*
                (sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict if x!=9)/
                sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict)))/
                sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict))

standard_errors_dict['NATIVE AMERICAN'] = np.sqrt(((native_american_speeding_dict[9]/sum(native_american_speeding_dict[x] for x in native_american_speeding_dict))*
                (sum(native_american_speeding_dict[x] for x in native_american_speeding_dict if x!=9)/
                sum(native_american_speeding_dict[x] for x in native_american_speeding_dict)))/
                sum(native_american_speeding_dict[x] for x in native_american_speeding_dict))

standard_errors_dict['OTHER'] = np.sqrt(((other_speeding_dict[9]/sum(other_speeding_dict[x] for x in other_speeding_dict))*
                (sum(other_speeding_dict[x] for x in other_speeding_dict if x!=9)/
                sum(other_speeding_dict[x] for x in other_speeding_dict)))/
                sum(other_speeding_dict[x] for x in other_speeding_dict))

standard_errors_dict['WHITE'] = np.sqrt(((white_speeding_dict[9]/sum(white_speeding_dict[x] for x in white_speeding_dict))*
                (sum(white_speeding_dict[x] for x in white_speeding_dict if x!=9)/
                sum(white_speeding_dict[x] for x in white_speeding_dict)))/
                sum(white_speeding_dict[x] for x in white_speeding_dict))
```

```python
all_races_speeding_dict = {}
all_races_length_dict = {}

all_races_speeding_dict['ASIAN'] = asian_speeding_dict
all_races_speeding_dict['BLACK'] = black_speeding_dict
all_races_speeding_dict['HISPANIC'] = hispanic_speeding_dict
all_races_speeding_dict['NATIVE AMERICAN'] = native_american_speeding_dict
all_races_speeding_dict['OTHER'] = other_speeding_dict
all_races_speeding_dict['WHITE'] = white_speeding_dict

for race in race_set:
    all_races_length_dict[race] = sum(all_races_speeding_dict[race][x] for x in all_races_speeding_dict[race])
```

```python
all_races_length_dict
```

```python
"""
Narrowing comparison to shares of Race's stops between 9-14 mph 

"""

total_stops_between_9_14 = {}
for x in race_set:
    total_stops_between_9_14[x] = 0
    
for x in range(9,15):
    total_stops_between_9_14['ASIAN'] += asian_speeding_dict[x]
    total_stops_between_9_14['BLACK'] += black_speeding_dict[x]
    total_stops_between_9_14['HISPANIC'] += hispanic_speeding_dict[x]
    total_stops_between_9_14['NATIVE AMERICAN'] += native_american_speeding_dict[x]
    total_stops_between_9_14['OTHER'] += other_speeding_dict[x]
    total_stops_between_9_14['WHITE'] += white_speeding_dict[x]

    
bernoulli_dict_9_14 = {}
bernoulli_dict_9_14['ASIAN'] = (asian_speeding_dict[9]/sum(asian_speeding_dict[x] for x in range(9,15)))
bernoulli_dict_9_14['BLACK'] = (black_speeding_dict[9]/sum(black_speeding_dict[x] for x in range(9,15)))
bernoulli_dict_9_14['HISPANIC'] = (hispanic_speeding_dict[9]/sum(hispanic_speeding_dict[x] for x in range(9,15)))
bernoulli_dict_9_14['NATIVE AMERICAN'] = (native_american_speeding_dict[9]/sum(native_american_speeding_dict[x] for x in range(9,15)))
bernoulli_dict_9_14['OTHER'] = (other_speeding_dict[9]/sum(other_speeding_dict[x] for x in range(9,15)))
bernoulli_dict_9_14['WHITE'] = (white_speeding_dict[9]/sum(white_speeding_dict[x] for x in range(9,15)))
```

RESULT #1:

t-stats comparing probability that Bernoulli variables $P(\frac{Cited @ 9 mph}{Cited})$ are equal
to the corresponding 'WHITE' Bernoulli variable
    
Reject this null for all races @ 95%


```python
bernoulli_dict = {}
bernoulli_dict['ASIAN'] = (asian_speeding_dict[9]/sum(asian_speeding_dict[x] for x in asian_speeding_dict))
bernoulli_dict['BLACK'] = (black_speeding_dict[9]/sum(black_speeding_dict[x] for x in black_speeding_dict))
bernoulli_dict['HISPANIC'] = (hispanic_speeding_dict[9]/sum(hispanic_speeding_dict[x] for x in hispanic_speeding_dict))
bernoulli_dict['NATIVE AMERICAN'] = (native_american_speeding_dict[9]/sum(native_american_speeding_dict[x] for x in native_american_speeding_dict))
bernoulli_dict['OTHER'] = (other_speeding_dict[9]/sum(other_speeding_dict[x] for x in other_speeding_dict))
bernoulli_dict['WHITE'] = (white_speeding_dict[9]/sum(white_speeding_dict[x] for x in white_speeding_dict))

z_stat_dict = {}
for race in race_set:
    if race!='WHITE':
        z_stat_dict[race] = (bernoulli_dict[race]-bernoulli_dict['WHITE'])/np.sqrt(((bernoulli_dict[race]*(1 - bernoulli_dict[race]))/all_races_length_dict[race])+((bernoulli_dict['WHITE']*(1 - bernoulli_dict['WHITE']))/all_races_length_dict['WHITE']))

z_stat_dict
```

RESULT #2 - robustness check:

t-stats comparing probability that Bernoulli variables $P(\frac{Cited @ 9 mph}{Cited \in \{9,14\} mph})$ are equal
to the corresponding 'WHITE' Bernoulli variable
    
Reject this null for all races except 'ASIAN' @ 95%


```python
"""
RESULT #2:

    t-stats comparing probability that Bernoulli variables P(Cited @ 9 mph / Cited in [9,14])


"""

z_stat_dict_9_14 = {}
for race in race_set:
    if race!='WHITE':
        z_stat_dict_9_14[race] = (bernoulli_dict_9_14[race]-bernoulli_dict_9_14['WHITE'])/np.sqrt(((bernoulli_dict_9_14[race]*(1 - bernoulli_dict_9_14[race]))/sum(all_races_speeding_dict[race][x] for x in range(9,15)))+((bernoulli_dict_9_14['WHITE']*(1 - bernoulli_dict_9_14['WHITE']))/sum(all_races_speeding_dict['WHITE'][x] for x in range(9,15))))

z_stat_dict_9_14
```

```python
n_groups = 1
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 1/9
opacity = 1

rects1 = plt.bar(index, bernoulli_dict_9_14["WHITE"], bar_width,
alpha=opacity,
label='White')

rects6 = plt.bar(index + 1/6, bernoulli_dict_9_14["ASIAN"], bar_width,
alpha=opacity,
label='Asian')

rects2 = plt.bar(index + 1/6*2, bernoulli_dict_9_14["OTHER"], bar_width,
alpha=opacity,
label='Other')

rects3 = plt.bar(index + 1/6*3, bernoulli_dict_9_14["HISPANIC"], bar_width,
alpha=opacity,
label='Hispanic')

rects4 = plt.bar(index + 1/6*4, bernoulli_dict_9_14["BLACK"], bar_width,
alpha=opacity,
label='Black')

rects5 = plt.bar(index + 1/6*5, bernoulli_dict_9_14["NATIVE AMERICAN"], bar_width,
alpha=opacity,
label='Native American')




plt.ylabel("Proportion")
plt.xticks([])
plt.title('% Adjusted Speeds out of Possible Adjusted Speeds')
plt.ylim(.9,.95)
plt.legend(loc=1, prop={'size': 7})
plt.grid(linestyle='dotted')
style.use('tableau-colorblind10')
# plt.savefig('ROBUST Speeding Analysis Figure.png',dpi=1200)
```
