# Functions

## Add H3 series to df

```python
def create_h3_series(dataframe:pd.DataFrame, resolution:int) -> pd.Series:
    return h3.geo_to_h3(lat=dataframe['Longitude'], 
                        lng=dataframe['Latitude'],
                        resolution=resolution)
```

```python
def add_h3_to_df(dataframe:pd.DataFrame, resolutions:list) -> pd.DataFrame:
    
    for res in resolutions:
        dataframe[f'H3 Encoding - Res={res}'] = dataframe.apply(create_h3_series, resolution=res, axis=1)
    
    return dataframe
```

## Final Fnc

```python
def create_encodings(df_filepaths:list, resolutions:list) -> None:
    """Create geo-hex encodings and then overwrites file - for one """
    for path in df_filepaths:
        add_h3_to_df(dataframe=pd.read_csv(clean_path(path)),
                     resolutions=resolutions).to_csv(path, index=False)
```
