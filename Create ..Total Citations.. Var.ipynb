```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"Traffic_Violations.csv")
```

```python
df.insert(0, 'Unique Stop ID', [x[0]+x[1]+x[2] for x in zip(df['SeqID'], df['Date Of Stop'], df['Time Of Stop'])])
del df['SeqID']
del df['Geolocation']
```

```python
citation_counting_dict = {ID: 0 for ID in df['Unique Stop ID']}

for i, ID in enumerate(df['Unique Stop ID']):
    if df['Violation Type'].iloc[i] == 'Citation':
        citation_counting_dict[ID] += 1    

df.insert(3, 'Number of Citations', [citation_counting_dict[ID] for ID in df['Unique Stop ID']])
```

```python
df = df.sort_values(by=['Date Of Stop', 'Time Of Stop', 'Unique Stop ID'])
```
